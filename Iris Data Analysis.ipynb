## Convert data into pandas dataframe

```python
df=pd.DataFrame(Iris.data,columns=Iris.feature_names)
```

```python
df.head()         #show first 5 values
```

```python
df.tail()        #shows last 5 values
```

```python
#lets include the target
df['target']=Iris.target
```

```python
df.head()
```

```python
df.dtypes         #display datatypes of each attributes
```

```python
df.describe()      #display statistical details
```

```python
print(df.groupby('target').size())    
```
